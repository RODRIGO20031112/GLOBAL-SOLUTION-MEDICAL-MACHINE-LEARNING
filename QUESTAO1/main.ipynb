```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

# Lê os dados do arquivo "Mall_Customers.csv" usando o Pandas
df = pd.read_csv("Breast_Cancer.csv")

# Seleciona as colunas 'Annual Income (k$)' e 'Spending Score (1-100)' como features para o agrupamento
X = df[['Tumor Size', 'Survival Months']]

# Padroniza as features usando StandardScaler para normalizá-las
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Lista para armazenar a inércia (soma dos quadrados das distâncias dentro de cada cluster) para diferentes valores de k
inertia = []

# Loop para testar diferentes valores de k (número de clusters) e calcular a inércia para cada k
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plotagem do gráfico "Elbow" para ajudar na escolha do número ótimo de clusters
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Inércia')
plt.title('Método Elbow para Escolha de k')
plt.grid(True)
plt.show()

# Após a escolha do número de clusters, aplica o K-Means novamente com k=5
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(X_scaled)

# Calcula e exibe o Silhouette Score, uma métrica de qualidade do agrupamento
silhouette_avg = silhouette_score(X_scaled, kmeans.labels_)
print(f'Silhouette Score: {silhouette_avg}')
```

```python
import subprocess
import sys
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from matplotlib.lines import Line2D

def install_libraries():
    libraries = ['pandas', 'matplotlib', 'sklearn']
    for library in libraries:
        try:
            __import__(library)
        except ImportError:
            subprocess.check_call([sys.executable, "-m", "pip", "install", library])

install_libraries()

def perform_pca_and_visualize(df, features, color_column):
    # Padroniza os dados (importante para o PCA)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(df[features])

    # Inicializa o PCA e ajuste aos dados
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_scaled)

    # Cria um DataFrame para visualizar os resultados do PCA
    df_pca = pd.DataFrame(data=X_pca, columns=['Main Component 1', 'Main Component 2'])

    # Define o mapa de cores
    if color_column == 'Race':
        # Se a coluna for 'Race', use diferentes cores para cada categoria de raça
        unique_races = df[color_column].unique()
        color_map = dict(zip(unique_races, plt.cm.get_cmap('tab10', len(unique_races)).colors))
        colors = df[color_column].map(color_map)
        legend_labels = [Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=label) 
                   for label, color in color_map.items()]
    else:
        # Se a coluna não for 'Race', converta os valores para códigos categóricos e use o colormap padrão
        colors = pd.Categorical(df[color_column]).codes
        legend_labels = None

    # Visualiza os resultados
    plt.figure(figsize=(8, 6))
    sc = plt.scatter(df_pca['Main Component 1'], df_pca['Main Component 2'], c=colors, cmap='viridis')

    # Adicione uma barra de cores
    plt.colorbar(sc, label=color_column)

    # Adicione a legenda das cores
    plt.legend(handles=legend_labels, title=color_column, loc='upper left')

    plt.xlabel('Main Component 1')
    plt.ylabel('Main Component 2')
    plt.title(f'PCA - Breast Cancer dataset ({color_column})')
    plt.show()

# Modify the main function
def main():
    # Carrega o conjunto de dados "Mall Customers" do CSV
    df = pd.read_csv('Breast_Cancer.csv')

    # Seleciona as características para aplicar o PCA
    features = ['Tumor Size', 'Survival Months']

    # Execute PCA e visualiza os resultados apenas para colunas presentes no dataframe
    columns_to_visualize = [col for col in ['Age', 'Race', 'Marital Status', 'T Stage', 'N Stage', '6th Stage', 'differentiate', 'Grade', 'A Stage', 'Tumor Size', 'Estrogen Status', 'Progesterone Status', 'Regional Node Examined', 'Reginol Node Positive', 'Survival Months', 'Status'] if col in df.columns]

    for column in columns_to_visualize:
        perform_pca_and_visualize(df, features, column)

if __name__ == "__main__":
    main()
```

```python
import subprocess
import sys
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from matplotlib.lines import Line2D

def install_libraries():
    libraries = ['pandas', 'matplotlib', 'sklearn']
    for library in libraries:
        try:
            __import__(library)
        except ImportError:
            subprocess.check_call([sys.executable, "-m", "pip", "install", library])

install_libraries()

def perform_pca_and_visualize(df, features):
    # Padroniza os dados (importante para o PCA)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(df[features])

    # Inicializa o PCA e ajuste aos dados
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X_scaled)

    # Aplique o KMeans diretamente nos dados reduzidos
    kmeans = KMeans(n_clusters=3, random_state=42)
    clusters = kmeans.fit_predict(X_pca)

    # Cria um DataFrame para visualizar os resultados do PCA e dos clusters
    df_pca = pd.DataFrame(data=X_pca, columns=['Main Component 1', 'Main Component 2'])
    df_pca['Cluster'] = clusters

    # Visualiza os resultados
    plt.figure(figsize=(8, 6))
    sc = plt.scatter(df_pca['Main Component 1'], df_pca['Main Component 2'], c=clusters, cmap='viridis')

    # Adicione uma barra de cores
    plt.colorbar(sc, label='Cluster')

    plt.xlabel('Main Component 1')
    plt.ylabel('Main Component 2')
    plt.title('PCA with KMeans Clustering')
    plt.show()

# Modify the main function
def main():
    # Carrega o conjunto de dados "Breast Cancer" do CSV
    df = pd.read_csv('Breast_Cancer.csv')

    # Seleciona as características para aplicar o PCA
    features = ['Tumor Size', 'Survival Months']

    # Execute PCA e visualiza os resultados
    perform_pca_and_visualize(df, features)

if __name__ == "__main__":
    main()
```

```python
# Importação das bibliotecas necessárias para resolução do exercício.
# O LabelEncoder não foi necessário neste caso, pois o conjunto de dados de câncer de mama já é fornecido com características numéricas. 
# Portanto, não é necessário converter as variáveis categóricas em numéricas.
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_breast_cancer
```

```python
# Carregando o conjunto de dados de câncer de mama e armazenando em uma variável.
data = load_breast_cancer()
# Mostrando os dados
data
```

```python
# Definição das features e do Target(dados de entrada e saída usados para treinar e testar o modelo de aprendizado de máquina.)
X = data.data  # Features
y = data.target  # Target 
```

```python
# Normalização dos dados utilizando o StandardScaler.
# Um processo de padronização para transformar os dados para que eles tenham uma média zero e um desvio padrão igual a 1.
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Dividindo os dados em conjuntos de treinamento
# X_treino e y_treino: dados para treinamento dos modelos (80% dos dados iniciais)
# X_teste e y_teste: dados para teste e avaliação dos modelos (20% dos dados iniciais)
# Assim como foi utilizado e discutido no exercício em sala, foi concluído que usar uma divisão de 80% para treinamento e 20% são valores padrões para esse cenário.
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.20)
```

```python
# Verifica e imprime as dimensões (número de linhas e colunas) dos conjuntos de treinamento (X_treino) e teste (X_teste). 
# Esta verificação é útil para garantir que os dados tenham sido divididos corretamente.
X_treino.shape, X_teste.shape
```

```python
# Redução de dimensionalidade com PCA
# Reduzindo o número de dimensões para 8.
p = 8
pca = PCA(n_components=p)
```

```python
# Aplicando o algoritmo PCA aos dados de treinamento e teste, reduzindo o número de características para 8.
X_treino_pca = pca.fit_transform(X_treino)
X_teste_pca = pca.transform(X_teste)
```

```python
# Verificando e imprimindo as novas dimensões após o uso do algoritmo PCA
X_treino_pca.shape, X_teste_pca.shape
```

```python
# Verificando a proporção da variância explicada por cada componente principal após a aplicação do PCA. 
pca.explained_variance_ratio_
```

```python
# Calculando a variância total explicada pelos componentes principais(Novas variáveis que o PCA criou após a redução de dimensionalidade) e imprimindo a porcentagem de variância explicada pelo modelo PCA.
total_variancia = pca.explained_variance_ratio_.sum()
print(f'O modelo PCA com {p} variáveis explica{100*(total_variancia): .2f} % dos dados de entrada')
```

```python
# Criando um modelo de classificação Random Forest com 40 estimadores (árvores de decisão) e o treinando com os dados de treinamento originais.
# Utilizei 40 pois tive como base o código do exercício de PCA que foi passado. 
modelo_rf = RandomForestClassifier(n_estimators=40, random_state=0)
modelo_rf.fit(X_treino, y_treino)
```

```python
# Usando o modelo Random Forest treinado para fazer previsões com base nos dados de teste originais.
previsoes = modelo_rf.predict(X_teste)
```
